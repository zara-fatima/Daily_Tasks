 Health Grade

```python
def health_grade(hr):
    if hr < 70:
        return "A (Excellent)"
    elif 70 <= hr <= 89:
        return "B (Stable)"
    elif 90 <= hr <= 110:
        return "C (Concern)"
    else:
        return "D (High Risk)"

df['Health_Grade'] = df['hr'].apply(health_grade)

df
```
