Task_2: Shopping cart management.

```python
cart=[]

while True:
    a=int(input("Enter '1' if you want to add item \n '2' to remove item \n '3' to update the price of an item \n '4' to exit the loop: \n "))
    
    # To add an item
    
    if a==1:
        _item=input("Enter the item you want to add: ")
        _price=int(input("Enter the price of the item: "))
        cart.append({"item":_item , "price":_price})
        
    #To remove an item
    
    elif a==2:
        removed_item=input("Enter the item you want to remove: ")
        for i in range(len(cart)):
            if cart[i]['item']==removed_item:
                cart.pop(i)
                break
    #To update the price of an item 

    elif a==3:
        _item=input("Enter the item you want to update the price of: ")
        _price=int(input("Enter the updated price of the item: "))
        for i in range(len(cart)):
            if cart[i]['item']==_item:
                cart[i]['price']=_price
    #Breaks the loop
    
    else:
        break
        
total_price=0

# Calculates and displays the total price

for i in range(len(cart)):
    total_price+=cart[i]['price']

print("Total price: ",total_price) 

# Search and displays the most expensive item

expensive_item=max(cart, key=lambda p:p['price'])
print('Most expensive item: ',expensive_item['item'], 'and its price: ',expensive_item['price'] )
       
```

Task 09: Sports Tournament Statistics.

```python
no_of_matches = int(input("Enter the number of matches: "))

teams = []        
scores = []       
participants = [] 

for i in range(no_of_matches):
    print(f"\nMatch {i+1}")
    team1 = input("Enter Team 1 name: ")
    score1 = int(input("Enter Team 1 score: "))
    team2 = input("Enter Team 2 name: ")
    score2 = int(input("Enter Team 2 score: "))

    players = int(input("Enter total players who played in this match: "))
    participants.append(players)

# Update Team 1
    if team1 in teams:
        index = teams.index(team1)
        scores[index] = scores[index] + score1
    else:
        teams.append(team1)
        scores.append(score1)

# Update Team 2
    if team2 in teams:
        index = teams.index(team2)
        scores[index] = scores[index] + score2
    else:
        teams.append(team2)
        scores.append(score2)

# find the winning team 
max_score = max(scores)
index = scores.index(max_score)
winner = teams[index]

# Average score
total_score = sum(scores)
average_score = total_score / len(scores)

# Total participants 
unique_participants = sum(participants)

#  Printing the required output 
print("\nTournament Summary")
for i in range(len(teams)):
    print(f"{teams[i]} total score: {scores[i]}")

print(f"\nWinning Team: {winner} ({max_score} points)")
print(f"Average Score per Team: {average_score}")
print(f"Total Participants: {unique_participants}")
```

Task 11: Employee Attendance System.

```python
employees = {}
num_employees = int(input("Enter number of employees: "))

# Input attendance for each employee
for i in range(num_employees):
    name = input(f"\nEnter name of employee {i+1}: ")
    attendance = input("Enter attendance for 30 days (P for present, A for absent): ").upper()
    employees[name] = attendance

perfect_attendance = []
most_absent_name = None
max_absences = -1

# Process each employee
for name, record in employees.items():
    presents = record.count('P')
    absences = record.count('A')
    percentage = (presents / 30) * 100

    print(f"{name}: {presents} Present, {absences} Absent, Attendance = {percentage:.2f}%")

    # Check for perfect attendance
    if absences == 0:
        perfect_attendance.append(name)

    # Track who has the most absences
    if absences > max_absences:
        max_absences = absences
        most_absent_name = name

# Printing the required output

if perfect_attendance:
    print("Perfect Attendance:", ", ".join(perfect_attendance))
else:
    print("No one had perfect attendance.")

print(f"Most Absences: {most_absent_name} ({max_absences} days)")
```

```python
users = {}
num_users = int(input("Enter number of users: "))

for i in range(num_users):
    name = input(f"\nEnter name of user {i+1}: ")
    num_activities = int(input(f"How many activities did {name} do? "))
    
    activities = []
    total_calories = 0
    total_time = 0

    for j in range(num_activities):
        activity = input(f"Enter activity {j+1} name: ").lower()
        time = float(input("Enter workout time in minutes: "))
        calories = float(input("Enter calories burned: "))
        print()

        activities.append(activity)
        total_calories += calories
        total_time += time

    users[name] = {
        "activities": activities,
        "total_calories": total_calories,
        "avg_time": total_time / num_activities
    }

# Printing the required output

all_activities = set()

for name, data in users.items():
    all_activities.update(data["activities"])
    print(f"\n{name}:")
    print(f" Total Calories Burned: {data['total_calories']:.2f}")
    print(f" Average Workout Time: {data['avg_time']:.2f} minutes")

print("\nUnique Activity Types:", ", ".join(sorted(all_activities)))
```

Task 13: Online Quiz Evaluator.

```python
students = {}
num_students = int(input("Enter number of students: "))

# Input data
for i in range(num_students):
    name = input(f"\nEnter name of student {i+1}: ")
    score = float(input("Enter quiz score: "))
    time_taken = float(input("Enter time taken (in minutes): "))

    students[name] = {"score": score, "time": time_taken}

# 1. Average score
total_score = 0
for student in students.values():
    total_score += student["score"]

average_score = total_score / num_students

# 2. Highest scorer
highest_scorer = max(students, key=lambda name: students[name]["score"])
highest_score = students[highest_scorer]["score"]

# 3. Passing threshold
min_score = float(input("\nEnter minimum score to pass: "))

passed_students = [] 
for name, data in students.items():
    if data["score"] >= min_score:
        passed_students.append(name) 

# Display results
print(f"\nAverage Score: {average_score:.2f}")
print(f"Highest Scorer: {highest_scorer} ({highest_score} marks)")

print("\nStudents who passed:")
if passed_students:
    for name in passed_students:
        print(f" - {name}")
else:
    print("No students passed.")
```

Task 14: Hotel Booking Management.

```python
# Rooms and their rates
rooms = {
    101: 100,
    102: 120,
    103: 150,
    104: 200,
    105: 180
}

occupied_rooms = []  # List to track booked rooms

while True:
    print("\nHotel Booking Menu")
    print("1. Book a room")
    print("2. View occupied rooms")
    print("3. View available rooms")
    print("4. Exit")
    
    choice = input("Enter your choice: ")

    if choice == "1":
        room_number = int(input("Enter room number to book: "))
        if room_number not in rooms:
            print("Invalid room number.")
        elif room_number in occupied_rooms:
            print(f"Room {room_number} is already booked.")
        else:
            nights = int(input("Enter number of nights: "))
            total_cost = nights * rooms[room_number]
            print(f"Room {room_number} booked for {nights} nights. Total cost: {total_cost}")
            occupied_rooms.append(room_number)

    elif choice == "2":
        if occupied_rooms:
            print("Occupied Rooms:", ", ".join(str(r) for r in occupied_rooms))
        else:
            print("No rooms are currently occupied.")

    elif choice == "3":
        available_rooms = [r for r in rooms if r not in occupied_rooms]
        if available_rooms:
            print("Available Rooms:", ", ".join(str(r) for r in available_rooms))
        else:
            print("No rooms are available.")

    elif choice == "4":
        print("Exited")
        break

    else:
        print("Invalid choice. Please enter 1, 2, 3, or 4.")
```

Task 15: Expense & Budget Tracker.
\

```python
budget = {}

num_categories = int(input("Enter number of expense categories: "))
for _ in range(num_categories):
    category = input("Enter category name: ")
    amount = float(input(f"Enter budget for {category}: "))
    budget[category] = amount

# Step 2: Record daily expenses
expenses = {}
while True:
    category = input("Enter expense category (or 'done' to finish): ")
    if category.lower() == 'done':
        break
    if category not in expenses:
        expenses[category] = []
    amount = float(input(f"Enter expense amount for {category}: "))
    expenses[category].append(amount)

# Step 3: Calculate total per category and check budget
print("\nExpense Summary:")
largest_expense = 0
largest_category = ""
for category, amounts in expenses.items():
    total = sum(amounts)
    print(f"{category}: Total = {total}")
    if category in budget:
        if total > budget[category]:
            print(f"You exceeded your budget in {category} by {total - budget[category]}")
    # Track largest expense
    max_expense = max(amounts)
    if max_expense > largest_expense:
        largest_expense = max_expense
        largest_category = category

# Step 4: Largest expense
print(f"\nLargest single expense: {largest_expense} in category '{largest_category}'")
```

Task 16: Integrated Real-World System: Smart Business Manager.

```python
transactions = []  

# Function to add a transaction
def add_transaction():
    date = input("Enter date (DD/MM/YYYY): ")
    t_type = input("Enter type (Income/Expense): ").capitalize()
    amount = float(input("Enter amount: "))
    category = input("Enter category: ").capitalize()

    # Store as dictionary
    transaction = {
        "date": date,
        "type": t_type,
        "amount": amount,
        "category": category
    }
    transactions.append(transaction)
    print("Transaction added successfully!\n")

# Function to show total summary by category
def total_summary():
    summary = {}
    for t in transactions:
        cat = t["category"]
        if cat not in summary:
            summary[cat] = 0
        # Expenses are counted as negative
        if t["type"] == "Expense":
            summary[cat] -= t["amount"]
        else:
            summary[cat] += t["amount"]
    print("\n Total Summary by Category")
    for cat, total in summary.items():
        print(f"{cat}: {total}")

# Function to find highest expense
def highest_expense():
    expenses = [t for t in transactions if t["type"] == "Expense"]
    if not expenses:
        print("No expenses recorded.\n")
        return
    highest = max(expenses, key=lambda x: x["amount"])
    print("--- Highest Expense ---")
    print(f"Date: {highest['date']}")
    print(f"Category: {highest['category']}")
    print(f"Amount: {highest['amount']}")

# Function to print full report
def financial_report():
    print("\n Financial Summary Report")
    print("{:<12} {:<10} {:<10} {:<15}".format("Date", "Type", "Amount", "Category"))
    print("-" * 50)
    for t in transactions:
        print("{:<12} {:<10} {:<10} {:<15}".format(t["date"], t["type"], t["amount"], t["category"]))
    print("-" * 50)
    total_summary()
    highest_expense()

# Main menu
def main():
    while True:
        print("1. Add Transaction")
        print("2. Show Total Summary")
        print("3. Show Highest Expense")
        print("4. Generate Financial Report")
        print("5. Exit")

        choice = input("Enter your choice: ")
        if choice == '1':
            add_transaction()
        elif choice == '2':
            total_summary()
        elif choice == '3':
            highest_expense()
        elif choice == '4':
            financial_report()
        elif choice == '5':
            print("Program exited")
            break
        else:
            print("Invalid choice. Try again.\n")

# To run the program
main()
```
